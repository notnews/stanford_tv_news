```python
from os import path
from glob import glob
import json
import pandas as pd
```

```python
# Output
# archive_id (to join internet archive), video_id, channel, show, people, year, date, screen time
```

```python
video_meta = pd.read_csv('videos.csv.gzip', nrows=10)
video_meta.columns
```

```python
video_meta = pd.read_csv('videos.csv.gzip', usecols=['id', 'date', 'channel', 'show', 'name'])
video_meta
```

```python
# rename colunms
video_meta.columns = ['video_id', 'date', 'channel', 'show', 'archive_id']
```

```python
video_meta
```

```python
people_screentime = []
for i, fn in enumerate(sorted(glob('json/screentime/*.json'))):
    name = path.basename(fn).replace('.json', '')
    with open(fn, "r") as json_file:
        data = json.load(json_file)
    for k in data:
        for v in data[k]:
            people_screentime.append({'people': name, 'year': k, 'video_id': v[0], 'screentime': v[1]})
    if i % 10 == 0:
        print(i, fn, len(people_screentime))
    #break
```

```python
df = pd.DataFrame(people_screentime)
df
```

```python
mdf = df.merge(video_meta)
mdf
```

```python
mdf.to_csv('people_screentime.csv.gz', compression='gzip', index=False)
```

```python
gdf = mdf.groupby(['people', 'year']).agg({'screentime': 'sum'}).reset_index()
gdf
```

```python
gdf['screentime_minutes'] = gdf.screentime / 60.0
```
