```python
import requests
import time
import json
from os import path
```

```python
r = requests.get(f'https://tvnews.stanford.edu/data/people.json?_={int(time.time() * 1000)}')
if r.status_code == 200:
    data = r.json()['data']
    with open("json/people.json", "w") as json_file:
        json.dump(data, json_file)
```

```python
for i, p in enumerate(data):
    name = p[0]
    print(i, name)
    fn = f"json/screentime/{name}.json"
    if path.exists(fn):
        continue
    r = requests.get(f'https://tvnews.stanford.edu/search?aggregate=year&detailed=true&end_date=2024-08-20&query=["name","{name}"]&start_date=2010-01-01')
    if r.status_code == 200:
        st_data = r.json()
        with open(fn, "w") as json_file:
            json.dump(st_data, json_file)
    else:
        print('ERROR')
    #break
```
