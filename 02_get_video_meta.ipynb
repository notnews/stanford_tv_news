```python
from os import path
from glob import glob
import json
import requests
import pandas as pd
```

```python
video = set()
for i, fn in enumerate(sorted(glob('json/screentime/*.json'))):
    with open(fn, "r") as json_file:
        # Step 2: Load the JSON data into a Python object (e.g., a dictionary or list)
        data = json.load(json_file)
    for k in data:
        for v in data[k]:
            video.add(v[0])
    if i % 10 == 0:
        print(i, fn, len(video))
```

```python
vl = list(video)
len(vl)
```

```python
for i in range(0, len(vl), 10):
    fn = f'json/video/{vl[i]}.json'
    if i % 1000 == 0:
        print(i, fn)
    if path.exists(fn):
        continue
    ids = ','.join([str(x) for x in vl[i:i+10]])
    url = f'https://tvnews.stanford.edu/search-videos?ids=[{ids}]'
    r = requests.get(url)
    if r.status_code == 200:
        data = r.json()
        with open(fn, "w") as json_file:
            json.dump(data, json_file)
    else:
        print('ERROR')    
    #if i > 1000:
    #    break
```

```python
video_metadata = []
for i, fn in enumerate(glob('json/video/*.json')):
    if i % 1000 == 0:
        print(i, fn)
    with open(fn, "r") as json_file:
        # Step 2: Load the JSON data into a Python object (e.g., a dictionary or list)
        data = json.load(json_file)    
    for d in data:
        video_metadata.append(d['metadata'])
```

```python
df = pd.DataFrame(video_metadata)
df
```

```python
df[['id', 'date', 'channel', 'show', 'name', 'width', 'height', 'fps', 'num_frames']].to_csv('videos.csv.gz', compression='gzip', index=False)
```
